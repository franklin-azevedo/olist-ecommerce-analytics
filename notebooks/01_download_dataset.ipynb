```python
# ==================================================
# 1. Setup - Imports & Project Paths
# ==================================================
from pathlib import Path
import shutil
import kagglehub

# Define the target directory for raw data ingestion
RAW_DIR = Path("../data/raw")
RAW_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# ==================================================
# 2. Ingestion Pipeline - Download & Verify Dataset
# ==================================================

# Download the latest version of the Olist dataset via KaggleHub API
cache_raw_path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
cache_path = Path(cache_raw_path)

# Iterate and copy all CSV files from local cache to the project's raw directory
copied_files = []
for file_path in cache_path.glob("*.csv"):
    shutil.copy(file_path, RAW_DIR)
    copied_files.append(file_path.name)

# --------------------------------------------------
# Executive Ingestion Report (Strict Layout Alignment)
# --------------------------------------------------
PANEL_WIDTH = 55
print("=" * PANEL_WIDTH)
print(f"{'DATA INGESTION REPORT':^{PANEL_WIDTH}}")
print("=" * PANEL_WIDTH)
print(f"{'Status:':<20}{'SUCCESS':>35}")
print(f"{'Destination:':<20}{str(RAW_DIR):>35}")
print(f"{'Total CSVs Copied:':<20}{len(copied_files):>35}")
print("-" * PANEL_WIDTH)
print("Copied Files & Sizes:")

# Dynamic loop to calculate actual file sizes from the raw directory
for file in sorted(copied_files):
    file_size_mb = (RAW_DIR / file).stat().st_size / (1024 * 1024)
    print(f" - {file:<38} ({file_size_mb:>6.2f} MB)")
print("=" * PANEL_WIDTH)
```
